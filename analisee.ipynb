```python
# Passo 1
import pandas as pd
tabela = pd.read_csv("cancelamentos.csv")
display(tabela)
# Passo 2
tabela = tabela.drop(columns="CustomerID")
display(tabela)
```

```python
# Passo 3
display(tabela.info())
tabela = tabela.dropna()
display(tabela.info())
```

```python
# Passo 4
display(tabela["cancelou"].value_counts())

display(tabela["cancelou"].value_counts(normalize=True))
```

```python
# Passo 5 
import plotly.express as px

for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelou")
    grafico.show()
# Criar os graficos utlizia muito espaço, não irei dar run nesta parte do codigo
# Ele cria graficos de histograma de todas as colunas em relação a coluna "cancelou"
```

```python
'''
Analisando os dados podemos ver :
 assinantes com mais de 50 anos ou mais sempre cancelam
 assinantes que ligaram 5 vezes ou mais para o callcenter sempre cancelam
 pagamento que atrasa 20 dias ou mais sempre cancelam
 assinantes mensais sempre cancelam
 assinantes que gastaram menos de ou igual a 500 reais sempre cancelam
 feito a analise vamos ver quanto a porcentagem de usuarios estao cancelando depois de corrigir esses erros
'''
# Podemos ter limgugem acessivel ao publico mais velho, ou algum beneficio voltado a esse publico
tabela = tabela[tabela["idade"] <= 50]
# Podemos colocar a situação do cliente em prioridade se ele ligar mais de 3 vezes para o cliente
tabela = tabela[tabela["ligacoes_callcenter"] <= 5]
# Avisos sobre o atraso de pagamento e facilidade de resolver
tabela = tabela[tabela["dias_atraso"] <= 20]
# Podemos dar beneficios a assinaturas anuais e trimestrais como descontos e garantias
tabela = tabela[tabela['assinatura']!="Monthly"]
# Podemos aumentar os benefios as primeiras compras
tabela = tabela[tabela["total_gasto"]>=500]

display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True))
# Fazendo isso reduzimos o cancelamento para 88%
```
